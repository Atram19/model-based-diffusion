# How Model-Based Diffusion Works

Model-based diffusion can be viewed as an optimization algorithm. 

Let's say you have a highly-nonconvex function $J(x)$ that you want to minimize: 

```python
def J(x):
    y = 1.0 * ((x**4 - 0.5) ** 2 + 0.0 * x)
    y = y + np.sin(2 * np.pi * 10.0 * x) * 0.04
    return y
xs = np.linspace(-1, 1, 200)
ys = J(xs)
plt.plot(xs, ys)
```

MBD first transform the problem into a sampling problem by define $p_0(x) = \exp(-J(x)/\lambda)$

```python
def get_p0(x):
    lambda_J = 0.02
    return np.exp(-J(x) / lambda_J)
plt.plot(xs, get_p0(xs))
```
